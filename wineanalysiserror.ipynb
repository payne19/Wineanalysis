```python
import pandas as pd
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv('wine.csv')
```

```python
data1 = data.dropna(axis=0)
```

```python
y = data1.quality
```

```python
y
```

```python
data1.columns
```

```python
features = ['free sulfur dioxide','pH','alcohol']
```

```python
x  = data1[features]
```

```python
train_x,val_x,train_y,val_y = train_test_split(x,y,random_state=1)
```

```python
model_check = DecisionTreeRegressor(random_state=1)
```

```python
model_check.fit(train_x,train_y)
```

```python
validity = model_check.predict(val_x)
```

```python
print(mean_absolute_error(val_y, validity))
```
